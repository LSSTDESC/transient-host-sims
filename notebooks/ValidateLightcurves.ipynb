```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from astropy import table
```

```python
#plotting functionality!!
sns.set_context("talk",font_scale=1.5)

sns.set_style('white', {'axes.linewidth': 0.5})
plt.rcParams['xtick.major.size'] = 15
plt.rcParams['ytick.major.size'] = 15

plt.rcParams['xtick.minor.size'] = 10
plt.rcParams['ytick.minor.size'] = 10
plt.rcParams['xtick.minor.width'] = 2
plt.rcParams['ytick.minor.width'] = 2

plt.rcParams['xtick.major.width'] = 2
plt.rcParams['ytick.major.width'] = 2
plt.rcParams['xtick.bottom'] = True
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.left'] = True
plt.rcParams['ytick.right'] = True

plt.rcParams['xtick.minor.visible'] = True
plt.rcParams['ytick.minor.visible'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

plt.rcParams.update({
    "text.usetex": False,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
## for Palatino and other serif fonts use:
plt.rcParams.update({
    "text.usetex": False,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

```python
transDir = "/global/cscratch1/sd/kessler/SNANA_LSST_SIM/SCOTCH_CATALOG/"
prefix = "MLAG_GP_SCOTCH_FINAL"

AGN20_phot_fn = transDir + prefix + "_AGN20/MLAG_GP_SCOTCH_FINAL_NONIaMODEL0-0015_PHOT.FITS.gz"
AGN20_head_fn = transDir + prefix + "_AGN20/MLAG_GP_SCOTCH_FINAL_NONIaMODEL0-0015_HEAD.FITS.gz"

AGN01_phot_fn  = transDir + prefix + "_AGN01/MLAG_GP_SCOTCH_FINAL_AGN01_NONIaMODEL0-0021_PHOT.FITS.gz"
AGN01_head_fn  = transDir + prefix + "_AGN01/MLAG_GP_SCOTCH_FINAL_AGN01_NONIaMODEL0-0021_HEAD.FITS.gz"

KN_B19_phot_fn  = transDir + prefix + "_KN_B19/MLAG_GP_SCOTCH_FINAL_KN_NONIaMODEL0-0012_PHOT.FITS.gz"
KN_B19_head_fn  = transDir + prefix + "_KN_B19/MLAG_GP_SCOTCH_FINAL_KN_NONIaMODEL0-0012_HEAD.FITS.gz"

KN_K17_phot_fn  = transDir + prefix + "_KN_K17/MLAG_GP_SCOTCH_FINAL_KN_NONIaMODEL0-0012_PHOT.FITS.gz"
KN_K17_head_fn  = transDir + prefix + "_KN_K17/MLAG_GP_SCOTCH_FINAL_KN_NONIaMODEL0-0012_HEAD.FITS.gz"
```

```python
TDE_phot_fn  = transDir + prefix + "_TDE/MLAG_GP_SCOTCH_FINAL_NONIaMODEL0-0012_PHOT.FITS.gz"
TDE_head_fn  = transDir + prefix + "_TDE/MLAG_GP_SCOTCH_FINAL_NONIaMODEL0-0012_HEAD.FITS.gz"

AGN20_phot = table.Table.read(AGN20_phot_fn, format='fits').to_pandas()
AGN20_head = table.Table.read(AGN20_head_fn, format='fits').to_pandas()

AGN01_phot = table.Table.read(AGN01_phot_fn, format='fits').to_pandas()
AGN01_head = table.Table.read(AGN01_head_fn, format='fits').to_pandas()

KN_B19_phot = table.Table.read(KN_B19_phot_fn, format='fits').to_pandas()
KN_B19_head = table.Table.read(KN_B19_head_fn, format='fits').to_pandas()

KN_K17_phot = table.Table.read(KN_K17_phot_fn, format='fits').to_pandas()
KN_K17_head = table.Table.read(KN_K17_head_fn, format='fits').to_pandas()

TDE_phot = table.Table.read(TDE_phot_fn, format='fits').to_pandas()
TDE_head = table.Table.read(TDE_head_fn, format='fits').to_pandas()
```

```python
NEPOCH = len(np.unique(AGN20_phot['MJD']))
NBAND = 6

plt.figure(figsize=(10,7))
for i in np.arange(5):
    N = (NBAND*(NEPOCH-1))
    oneLC = AGN20_phot.iloc[N*i:N*(i+1)]
    oneLC_u = oneLC[oneLC['BAND'] == b'u ']
    oneLC_g = oneLC[oneLC['BAND'] == b'g ']
    oneLC_r = oneLC[oneLC['BAND'] == b'r ']
    oneLC_i = oneLC[oneLC['BAND'] == b'i ']
    oneLC_z = oneLC[oneLC['BAND'] == b'z ']
    oneLC_Y = oneLC[oneLC['BAND'] == b'Y ']
    plt.plot(oneLC_i['MJD'], oneLC_i['SIM_MAGOBS'], 'o', ms=1)
plt.gca().invert_yaxis()
```

```python
#the master plot with all of them together: 
NEPOCH_B19 = len(np.unique(KN_B19_phot['MJD']))
NEPOCH_K17 = len(np.unique(KN_K17_phot['MJD']))
NEPOCH_AGN01 = len(np.unique(AGN01_phot['MJD']))
NEPOCH_AGN20 = len(np.unique(AGN20_phot['MJD']))
NEPOCH_TDE = len(np.unique(TDE_phot['MJD']))

NEPOCHS = [NEPOCH_B19, NEPOCH_K17, NEPOCH_AGN01, NEPOCH_AGN20, NEPOCH_TDE]
phot_dfs = [KN_B19_phot, KN_K17_phot, AGN01_phot, AGN20_phot, TDE_phot]
classes = ["KN (Bulla+2019)", "KN (Kasen+2017)", r"AGN ($0.1d$)", r"AGN ($2.0d$)", 'TDE']

NBAND = 6
ms = [5, 5, 1, 5, 5]
peak = [53110, 53110, 53200, 53200, 53200]
markers = ['d', 'o','o','o','o']
cols = ['tab:orange', 'tab:orange', 'tab:green','tab:red','tab:blue']
plt.figure(figsize=(10,7))
for i in np.arange(len(ms)):
    N = (NBAND*(NEPOCHS[i]-1))
    oneLC = phot_dfs[i].iloc[N*i:N*(i+1)]
    oneLC_i = oneLC[oneLC['BAND'] == b'i ']
    plt.plot(oneLC_i['MJD']-peak[i], oneLC_i['SIM_MAGOBS'], markers[i], ms=ms[i], c=cols[i], label=classes[i])
plt.xlabel("Phase from Peak (Days)")
plt.ylabel(r"$m_i$")
plt.legend(loc='lower right', fontsize=16)
plt.ylim((35, 20))
plt.xlim((-10, 100))
plt.savefig("../plots/LC_Sampling.png",dpi=200, bbox_inches='tight')
```
